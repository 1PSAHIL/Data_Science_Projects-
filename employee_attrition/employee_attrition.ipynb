```python
# One-hot encode categorical variables
from category_encoders import OneHotEncoder
x=df.drop(['Attrition'],axis='columns')
y=df.Attrition
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
# Build a pipeline for one-hot encoding and logistic regression
model=make_pipeline(
    OneHotEncoder(),
    LogisticRegression()
)
model.fit(x_train,y_train)
```

```python
len(df.Attrition.unique())
```

```python
model.score(x_test,y_test) #Model Pridict 88 times right
```

```python
features = model.named_steps["onehotencoder"].get_feature_names_out()
importance=model.named_steps['logisticregression'].coef_[0]
features
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix
ConfusionMatrixDisplay.from_estimator(model, x_test, y_test);
```

```python
probilities = model.predict_proba(x_test)[:,1]
default_predictions=(probilities>=0.5).astype(int)
default_cm=confusion_matrix(y_test,default_predictions)
print("Confusion_matrix(DefaultThreshold 0.5)")
print(default_cm)
adjusted_threshold=0.7
adjusted_predictions=(probilities>=adjusted_threshold).astype(int)
adjusted_cm=confusion_matrix(y_test,adjusted_predictions)
print("\nConfusion_matrix(Adjusted Threshold 0.7)")
print(adjusted_cm)
```

```python
   #gini importance /odd ratios
odds_ratios=pd.Series(np.exp(importance),index=features).sort_values()
odds_ratios.head()
```

```python
odds_ratios.tail().plot(kind="barh")
```
